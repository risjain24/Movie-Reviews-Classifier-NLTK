```python
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
import string
stops = set(stopwords.words('english'))
stops.update(list(string.punctuation))
stops
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w,pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
# documments = [(clean_review(document), category) for document,category in documents]
documents_output = []
for document,category in documents:
    documents_output.append((clean_review(document),category))
```

```python
documents = documents_output
```

```python
documents[0][0]
```

```python
training_documents = documents[0:1500]
testing_documents = documents[1500:]
```

```python
all_words = []
for doc in training_documents:
    all_words += doc[0]
```

```python
import nltk
```

```python
freq = nltk.FreqDist(all_words)
features = [i[0] for i in freq.most_common(3000)]
features
```

```python
def get_feature_dict(words):
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
get_feature_dict(training_documents[0][0])
```

```python
training_data = []
for doc, category in training_documents:
    training_data.append((get_feature_dict(doc), category))
```

```python
training_data[0]
```

```python
testing_data = []
for doc, category in testing_documents:
    testing_data.append((get_feature_dict(doc), category))
```

```python
testing_data[0]
```

```python
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier, testing_data)
```
